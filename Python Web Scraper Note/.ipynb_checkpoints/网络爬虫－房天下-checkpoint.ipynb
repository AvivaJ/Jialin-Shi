```python
import requests
url = 'http://sh.esf.fang.com/'
domain = 'http://sh.esf.fang.com'
html= requests.get(url)
from bs4 import BeautifulSoup
soup = BeautifulSoup(html.text,'html.parser')
info = soup.select('dd h4 a')
for house in info:
    print(domain+house['href'])
    print('==========================')
```

```python
def getHouseInfo(url):
    html = requests.get(url)
    soup =BeautifulSoup(html.text,'html.parser')
    print(soup.select('h1')[0].text.strip())
    print(soup.select('.sty1')[0].text)
    [print(tt.text) for tt in soup.select('.tt')]
    
    
    
```

```python
import requests
url = 'http://sh.esf.fang.com/'
domain = 'http://sh.esf.fang.com'
html= requests.get(url)
from bs4 import BeautifulSoup
soup = BeautifulSoup(html.text,'html.parser')
info = soup.select('dd h4 a')
houseary= []
for house in info:
    url= domain+house['href']
    houseary.append(getHouseInfo(url))
    print('==========================')
```

```python
import pandas as pd
df =pd.DataFrame(houseary)
df
```
