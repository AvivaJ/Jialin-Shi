# requests库 爬取html源码

```python
import requests
url='https://movie.douban.com/top250'
headers={
    'Referer': 'https://www.google.com/'
    ,'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/83.0.4103.116 Safari/537.36'    
}

resp=requests.get(url,headers=headers)
text=resp.content.decode('utf-8')       #str格式的html
```

# lxml结合xpath语法 抽取数据

```python
from lxml import etree
```

##  把str解析成html 

```python
htmlElement=etree.HTML(text)                                       #这是一个element对象
etree.tostring(htmlElement,encoding='utf-8').decode('utf-8')       #转换成html的可读的str代码,用来看xpath
```

```python
movielist = htmlElement.xpath('//div[@class="item"]') # element.xpath()方
print(etree.tostring(movielist[1],encoding='utf-8').decode('utf-8'))
```

## xpath定位

```python
import pandas as pd 

movielist = htmlElement.xpath('//div[@class="item"]') # element.xpath()方法
posts=[]
for movie in movielist:
    try:
        link = movie.xpath('.//a/@href')[0]
        name = movie.xpath('.//span[@class="title"]/text()')[0]
        img_url = movie.xpath('.//img/@src')[0]
        cast= movie.xpath('.//p[@class=""]/text()')[0].replace('\n','').strip()
        tag= movie.xpath('.//p[@class=""]/text()')[1].replace('\n','').strip()
        score=movie.xpath('.//span[@class="rating_num"]/text()')[0]
        post={
            'link':link
            ,'name':name
            ,'img_url':img_url
            ,'cast':cast
            ,'tag':tag
            ,'score':score
        }
        posts.append(post)
    except:
        pass

df=pd.DataFrame(posts)
#     print(etree.tostring(link,encoding='utf-8').decode('utf-8'))
    
```

## 列表转成json储存

```python
import json

with open('豆瓣电影.json','w',encoding='utf-8') as fp:
    fp.write(json.dumps(posts))

with open('豆瓣电影.json','r',encoding='utf-8') as fp:
    data=json.load(fp)
```
